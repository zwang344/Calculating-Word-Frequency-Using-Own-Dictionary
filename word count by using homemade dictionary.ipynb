```python
import os
import glob
import pandas as pd
```

Open All Files

```python
path = 'D:\\Fordham\\Fashion AI\\SHFW\\reviewlist'
```

```python
reviewlist = []
for eachfile in glob.glob(os.path.join(path, '*.csv')):
    f = pd.read_csv(eachfile)
    f = f['reviews']
    
    content =''
    for line in f:
        content = content+line
    reviewlist.append(content)
```

Create A List For Our Own Dictionary

```python
dict_path = 'D:\\Fordham\\Fashion AI\\SHFW\\Fashion_Dictionary_new.csv'
```

```python
dictionary = pd.read_csv(dict_path, header=None, encoding='utf-8-sig')

dict_list = []
for word in dictionary[0]:
    dict_list.append(word)
```

Count Word Frequency for Each File

```python
wordcount_list=[]
for review in reviewlist:
    word_count={}
    for word in dict_list:
        word_count[word] = review.count(word)
    wordcount_list.append(word_count)
```

```python
rowname = ['2015AW', '2016AW','2016SS','2017AW','2017SS','2018AW','2018SS']
```

```python
wordcount_matrix = pd.DataFrame(wordcount_list, index=rowname)
wordcount_matrix
```

SAVE IN A CSV FILE

```python
wordcount_matrix.to_csv('D:\\Fordham\\Fashion AI\\SHFW\\wordcount_matrix.csv',  encoding='utf-8-sig')
```
